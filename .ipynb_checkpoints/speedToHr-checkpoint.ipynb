```python
import json
import os 
import fnmatch
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# import tensorflow as tf
# from tensorflow import keras
# from tensorflow.keras import layers
# from tensorflow.keras.layers.experimental import preprocessing
```

```python
## read data object
class Data:
    def __init__(self):
        self.workouts = []
        self.users = []

    def findFile(self,rootdir):
        fileList = os.listdir(rootdir)
        fldFileList = fnmatch.filter(fileList, '*.user.json')
        for i in range(len(fldFileList)):
            fldFileList[i] = fldFileList[i][0:-10]
        return fldFileList


    def findData(self,rootdir,fileName):
        with open(rootdir+"extf/"+fileName+".xx","r") as reader:
            self.workouts.append(json.loads(reader.read()))
        with open(rootdir+fileName+".user.json","r") as reader:
            self.users.append(json.loads(reader.read()))
```

```python
## collect the data
## jsonload folder include .user.json /extf/.xx
# rootdir = "/usr/app/Data/coros_data/jsonload/"
cwd = os.getcwd()
rootdir = cwd +"/data/"
data = Data()
fileNamelist = data.findFile(rootdir)
for i in range(len(fileNamelist)):
    data.findData(rootdir,fileNamelist[i])
```

```python
## select the data
i = 0
hr = np.zeros([len(data.workouts[i]),1])
sp = np.zeros([len(data.workouts[i]),1])

for j in range(len(data.workouts[i])):
    hr[j] = data.workouts[i][j]['hr']
    sp[j] = data.workouts[i][j]['sp']

startIndex = 0
endIndex = -1
fig, ax1 = plt.subplots()
ax1.plot(range(hr[startIndex:endIndex,:].shape[0]),hr[startIndex:endIndex,:],color='red',label='hr')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(range(sp[startIndex:endIndex,:].shape[0]),sp[startIndex:endIndex,:],color='blue',label='speed')
fig.legend()

```

```python
## cut the data (sp to sp2)
startIndex = 0
endIndex = 4000
dataAmount = endIndex - startIndex
sp2 = np.zeros([dataAmount,1])
sp2 = sp[startIndex:endIndex,0]
print(sp2.shape[0])
```

```python
class predict:
    def predicF(self,dataDiffIndex,b,gain,sp2,y,hr,startIndex,endIndex):
        y = np.zeros([sp2.shape[0]-dataDiffIndex,1])
        for i in range(y.shape[0]):
            y[i] = hr[startIndex]
        for i in range(sp2.shape[0]-dataDiffIndex):
            spDiff = sp2[i+dataDiffIndex] - sp2[i]
            if spDiff > 0:
                index = 0
            elif spDiff < 0:
                index = 1
            else:
                continue
            
            if sp2[i] <= self.thresholdTable[0]:
                index_x = 0
            elif sp2[i] < self.thresholdTable[1] and sp2[i] > self.thresholdTable[0]:
                index_x = 1
            elif sp2[i] < self.thresholdTable[2] and sp2[i] > self.thresholdTable[1]:
                index_x = 2
            elif sp2[i] < self.thresholdTable[3] and sp2[i] > self.thresholdTable[2]:
                index_x = 3
            elif sp2[i] < self.thresholdTable[4] and sp2[i] > self.thresholdTable[3]:
                index_x = 4
            else:
                index_x = 5
                    
            tau = self.tauTable[index_x,index]
            k = self.kTable[index_x,index]
            
            if self.predMethod == 1:
                predScale = sp2.shape[0]-dataDiffIndex
            else:
                predScale = self.predMethod
                
            for t in range(predScale):
                if t+i < sp2.shape[0]-dataDiffIndex:
                    y[i+t] += self.transferFunction(spDiff,gain,k,t,tau)
        return y

    def predicOfError(self,dataDiffIndex,b,gain,sp2,y,hr,startIndex,endIndex):
        y = self.predicF(dataDiffIndex,b,gain,sp2,y,hr,startIndex,endIndex)
        err = np.zeros([sp2.shape[0]-dataDiffIndex,1])
        err = y - hr[startIndex:endIndex-dataDiffIndex]
        return (np.dot(err.T,err)/err.shape[0])

    def __init__(self):
        ## model parameter
        self.tauTable = np.zeros([6,2])
        self.tauTable[0,0] = 9.4818
        self.tauTable[1,0] = 10.193
        self.tauTable[2,0] = 13.659
        self.tauTable[3,0] = 18.618
        self.tauTable[4,0] = 38.192
        self.tauTable[5,0] = 55.974

        self.tauTable[0,1] = 27.358
        self.tauTable[1,1] = 27.365
        self.tauTable[2,1] = 26.741
        self.tauTable[3,1] = 30.865
        self.tauTable[4,1] = 48.114
        self.tauTable[5,1] = 81.693

        self.kTable = np.zeros([6,2])
        self.kTable[0,0] = 9.2583
        self.kTable[1,0] = 11.264
        self.kTable[2,0] = 10.006
        self.kTable[3,0] = 12.807
        self.kTable[4,0] = 17.753
        self.kTable[5,0] = 32.911

        self.kTable[0,1] = 7.9297
        self.kTable[1,1] = 9.8561
        self.kTable[2,1] = 8.9772
        self.kTable[3,1] = 12.087
        self.kTable[4,1] = 12.087
        self.kTable[5,1] = 25.733

        self.thresholdTable = np.zeros([6,1])
        self.thresholdTable[0] = 0.5
        self.thresholdTable[1] = 1.5
        self.thresholdTable[2] = 2.0
        self.thresholdTable[3] = 2.5
        self.thresholdTable[4] = 3.0
        self.thresholdTable[5] = 3.5
        
        self.transfunction = 'stepResponse'
        self.predMethod = 1
        
    def transferFunction(self,spDiff,gain,k,t,tau):
        if self.transfunction == 'stepResponse':
            return spDiff*gain*(k-k*np.exp(-t/tau))
        elif self.transfunction == 'impulseResponse':
            return (spDiff*gain*(k/tau*np.exp(-t/tau)))
        else:
            return 0
        
    
```

```python
p = predict()
dataDiffIndex  = 4
gain = 1.0
p.transfunction = 'stepResponse'
p.predMethod = 1
y = np.zeros([dataAmount-dataDiffIndex,1])
y = p.predicF(dataDiffIndex,0,gain,sp2,y,hr,startIndex,endIndex)
plt.plot(y,label='predict HR')
plt.plot(hr[startIndex:endIndex-dataDiffIndex],label='true HR')
plt.legend()
```

## Following is using Ordinary least squares 

```python
## Ordinary least squares y(k) = a1 y(k-1) + a2 y(k-2) + ... + a10 y(k-10) + b0 u(k) + b1 u(k-1) + ... + b10 u(k-10)
## list of data to np.array of data
i = 0
lenOfData = len(data.workouts[i])-10
y = np.zeros([lenOfData,1])
a = np.zeros([lenOfData,12])
for j in range(len(data.workouts[i])-10):
    y[j][0] = data.workouts[i][j+10]['hr']
    a[j][0] = data.workouts[i][j+9]['hr']
    for k in range(11):
        a[j][k+1] = data.workouts[i][j+10-k]['sp']

## Ordinary least squares
ata = np.dot(a.transpose(),a)
ataInverse = np.linalg.inv(ata)
aty = np.dot(a.transpose(),y)
x = np.dot(ataInverse,aty)
print(x)
```
